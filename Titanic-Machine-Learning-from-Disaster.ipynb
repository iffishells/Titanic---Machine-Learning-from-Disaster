# Explore the Datasets Filling Missing Value

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("train.csv")

# data = data.head(30)

# data.describe() ## age and cabin and embarked must have some missing value 
# data.info()

print("Before FillNA :\n",data.isna().sum())

## drop is not good option because if we do this we loose all other information
# data.dropna(axis=0 ,how = "any" , inplace=True
data["Age"].fillna(axis=0,method = "ffill" ,limit=5 , inplace=True) ## forward fill 
data["Cabin"].fillna(axis=0,method = "bfill" ,limit=20, inplace=True) ## backward fill
print("Missing Value:\n",data.isna().sum())
#
data.dropna(axis=1,how="any",inplace=True)
```
